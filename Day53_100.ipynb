Your video game inventory system should:

    Have a menu that allows the user to:
        Add
        View
        Remove
    Adding an item saves it to a file using captalize mode. Duplicates are allowed.
    Removing an item deletes it from the file.
    View is trickier. It should output the name of the item and tell you how many of those items you have.
    Use auto-save and auto-load with try... except.

```python
import os, time

inventory = []
try:
  f = open("inventory.txt", "r")
  inventory = eval(f.read())
except:
  print("ERROR: No existing inventory list, using a blank list")


def add(item):

  inventory.append(item)
  print(f"{item} has been added to your inventory.")
  time.sleep(2)


def remove(item):
  if item in inventory:
    inventory.remove(item)
    print(f"{item} has been removed from your inventory.")
    time.sleep(2)


def view():
  temp = []
  for invent in inventory:
    count = 0

    for i in range(len(inventory)):
      if invent not in temp:
        if invent == inventory[i]:
          count += 1
    if invent not in temp:
      print(f"You {count:^2} {invent}")
    temp.append(invent)
    time.sleep(2)


while (True):
  os.system("clear")
  time.sleep(1)
  print(" ðŸŒŸRPG InventoryðŸŒŸ")
  print()
  user = input("1: Add  2: Remove  3: View \n >")
  print()
  try:
    if user == "1":

      item = input("Input the item to add: ").capitalize()
      add(item)
    elif user == "2":
      item = input("Input the item to add: ").capitalize()
      remove(item)
    elif user == "3":
      view()
  except:
    print(" Wrong Input. Try again")

  f = open("inventory.txt", "w")
  f.write(str(inventory))
  f.close()
```
