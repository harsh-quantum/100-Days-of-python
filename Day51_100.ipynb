```python
organizer_list = []
print("ðŸŒŸLife OrganizerðŸŒŸ\n")
print()

file  = open("oraganizer.txt", "r")
organizer_list = eval(file.read())
file.close()


def preetyprint():
    print()
    m =" "
    for row in organizer_list:
        print(f"Task : {row[0]:^10}{m:^2}|{m:^2} Date: {row[1]: ^10}{m:^2}|{m:^2} Priority: {row[2]:^10}")
    print()

def preety_prio_print(prior):
    print()
    m =" "
    for row in organizer_list:
        if prior in row[2]:
            print(f"Task : {row[0]:^10}{m:^2}|{m:^2} Date: {row[1]: ^10}{m:^2}|{m:^2} Priority: {row[2]:^10}")
    print()
    
def add_list():
    task = input("What is the task? >")
    due = input("When is it due by? >")
    priority = input("What is the priority? >")

    row = [task, due, priority]
    organizer_list.append(row)

    print()
    print("Thanks, this task has been added.")
    preetyprint()
    
def edit_list(edit_i):
    for row in range(len(organizer_list)):
        
        if edit_i in organizer_list[row][0]:
            
            organizer_list[row][0] = input("What is the task? >")
            organizer_list[row][1] = input("When is it due by? >")
            organizer_list[row][2] = input("What is the priority? >")
            print()
            print("Thanks, this task has been edited.")
            break
    preetyprint()
    
    
    
while(True): 
    print()
    user_i= input("Welcome to the life organizer. Do you want to add, view, edit or remove a to do? > ").strip().lower()
    if user_i == "add":
        add_list()
        
    elif user_i =="view":
        view_i = input("1. View all \n2. View priority \n ")
        if view_i == "1":
            preetyprint()
            
        elif view_i == "2":
            prior = input(" Enter the priority (high\medium\low) >  ").strip().lower()
            preety_prio_print(prior)

    elif user_i == "edit":
        preetyprint()
        edit_i = input("What do you want to edit? >")
        edit_list(edit_i)
            
    elif user_i == "remove":
        preetyprint()
        remove_i = input("What do you want to remove? >")
        for row in organizer_list:
            if remove_i in row:
                organizer_list.remove(row)
        print()
        print("Thanks, this task has been removed.")
        preetyprint()
    else:
        print("You provided the wrong input")


    x = input("Do you want to quit(y/n) ? ").strip().lower()
    if x[0] == "y":
        break

file = open("oraganizer.txt", "w")
file.write(str(organizer_list))
file.close()
```
