```python
import tkinter as tk
import re

cal = tk.Tk()
cal.title("Calculator")
cal.geometry("250x200")

answer = 0
operator = None
last = 0


def updatelabel_h(num):
    global answer
    answer = f"{answer}{num}"
    answer  = int(answer)
    screen["text"] = answer

def operation(op):
    
    global answer, last, operator
    last = answer
    answer = 0
    if op == "+" :
        operator = "+"
        
    elif op == "-" :
        operator = "-"
      
    elif op == "*" :
        operator = "*"
        
    else: 
    
        operator = "/"
    
    
   
    screen["text"] = answer
    
    
def finalresult():    
    global answer, last, operator
    total = 0
    if operator == "+" :
        total = answer + last
        
    elif operator == "-" :
        total = last - answer
        
    elif operator == "*" :
        total = answer * last    
    else: 
        total = last / answer 
    answer = total    
    screen["text"] = total
    
    
screen = tk.Label(text = answer)
# 'pack' places the element on the screen
screen.place(x=40, y=10)    
    
# Create a Button widget
b_1 = tk.Button(cal,text="1",command = lambda:updatelabel_h(1))
b_1.place(x=10, y=50)

b_2 = tk.Button(cal,text="2",command = lambda:updatelabel_h(2))
b_2.place(x=50, y=50)

b_3 = tk.Button(cal,text="3",command = lambda:updatelabel_h(3))
b_3.place(x=90, y=50)

b_4 = tk.Button(cal,text="4",command = lambda:updatelabel_h(4))
b_4.place(x=10, y=80)

b_5 = tk.Button( cal,text="5",command = lambda:updatelabel_h(5))
b_5.place(x=50, y=80)

b_6 = tk.Button(cal,text="6",command = lambda:updatelabel_h(6))
b_6.place(x=90, y=80)

b_7 = tk.Button(cal, text="7",command = lambda:updatelabel_h(7))
b_7.place(x=10, y=110)

b_8 = tk.Button(cal,text="8",command = lambda:updatelabel_h(8))
b_8.place(x=50, y=110)

b_9 = tk.Button(cal, text="9",command = lambda:updatelabel_h(9))
b_9.place(x=90, y=110)

b_0 = tk.Button(cal,text="0",command = lambda:updatelabel_h(0))
b_0.place(x=50, y=140)


b_plus = tk.Button(cal,text="+",command = lambda:operation('+'))
b_plus.place(x=130, y=50)

b_minus = tk.Button(cal,text="-",command = lambda:operation('-'))
b_minus.place(x=170, y=50)

b_mul = tk.Button(cal,text="*",command = lambda:operation('*'))
b_mul.place(x=130, y=80)

b_div = tk.Button(cal,text="/",command = lambda:operation('/'))
b_div.place(x=170, y=80)

b_eq = tk.Button(cal,text="=",command = finalresult)
b_eq.place(x=130, y=140)



# Start the Tkinter event loop
tk.mainloop()
```
