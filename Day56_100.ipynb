I've included a csv file ('100MostStreamedSongs.csv') in this repl that contains data from a music streaming service.

Your program should:

    Read in the data.
    Categorize the songs by artist, like this:
        Create one empty folder per artist.
        Create one blank text file per song by that artist in the relevant folder. The file name of the text file should be the name of the song.

```python
import csv
import os

    
    
        

with open("100MostStreamedSongs.csv") as file:

    reader = csv.DictReader(file)
    count = 0
    for row in reader:
        # i created one backup folder then did all the functionality
        
        listitems = os.listdir("backup")
        artist = row["Artist(s)"].title()
        print(artist)
        song = row["Song"]
        
        if artist not in listitems:
            os.mkdir(f"backup/{artist}")
            
        # Join the sanitized file name with the directory path
        sanitized_file_name = ''.join(c if c.isalnum() or c in {'_', '.', '-'} else '_' for c in song)
        
        # Now you can use the sanitized file path for your file operations
        file_path = os.path.join('your_directory_path', sanitized_file_name)
        
        path = os.path.join(f"backup/{artist}/",sanitized_file_name)
        
        
        print("Sanitized File Path:", file_path)

        
        file = open(path,"w")
        print(file)
        file.close()
       
      
```
