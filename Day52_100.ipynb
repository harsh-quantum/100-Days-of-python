That's right, you're opening a pizza shop! Try not to get anchovy on your keyboard. Man that stuff never washes out.

Regardless, your program today must:

    Prompt the user to input the quantity and size of pizzas.

    Multiply the two inputs together to calculate the cost of the pizzas.

    Store that in a 2D list with the user's name.

    Use try.... except for two reasons:
        Include auto-save and auto-load. Use it with the auto-load.
        When casting the quantity of pizzas to an integer. Avoid the user crashing the program by typing 'three' instead of '3'. Or any other non-integer input. If they do, then prompt them to try again.

```python
import time , os 
pizza_detail=[]

try:
  f = open("pizza.txt","r")
  pizza_detail = eval(f.read())
  f.close()
except:
  print(" ERROR: No existing file")

def add():
  name = input(" Name of the pizza: ")
  while(True):
    try: 
      quantity = int(input("How many pizzas? "))
      break
    except :
      print("You must input a numerical character,")
      print("try again")
  
    
    
  size = input("What size( s\m\l) ? ").lower()
  if  size=='s':
    price = 5
  elif size=="m":
    price = 10
  else:
    price = 15

  total = price*quantity
  row = [ name, quantity, size, total]
  pizza_detail.append(row)

def view(): 
  x =" "
  for row in pizza_detail:
    print(f"Name:{row[0]:^10}{x:^3}|{x:^3}Quantity:{row[1]:^10}{x:^3}|{x:^3}Size:{row[2]:^10}{x:^3}|{x:^3}Total:{row[3]:^10}{x:^3}")

while(True):
  time.sleep(1)
  os.system("clear")
  print("Welcome to the Menu")
  print("1. Add ")
  print("2. View")
  user = input()
  
  if user == "1":
    add()
  elif user == "2":
    view()  
  else:
    break
f= open("pizza.txt","w")
f.write(str(pizza_detail))
f.close()
```
