Your system should:

    Have a menu that asks if you want to add, view, move or edit a 'to do'.

    If you choose 'add' then the system should:
        Prompt you to input what the to do is, when it is due by and the priority (high, medium or low).
        Add the 'to do' to the list.

    'View' should give two options:
        View all - shows all 'to dos' with a pretty print.
        View priority - allows you to search for high, medium or low priority and only see matching tasks.

    'Edit' allows you to change any of the information within one of the 'to dos'.

    'Remove' lets you completely remove a 'to do' when it is 'to done'.

```python





organizer_list = []
print("ðŸŒŸLife OrganizerðŸŒŸ\n")
print()
while(True): 
  print()
  user_i= input("Welcome to the life organizer. Do you want to add, view, edit or remove a to do? > ").strip().lower()
  if user_i == "add":
    task = input("What is the task? >")
    due = input("When is it due by? >")
    priority = input("What is the priority? >")
    
    row = [task, due, priority]
    organizer_list.append(row)
    print()
    print("Thanks, this task has been added.")
  
  elif user_i =="view":
    view_i = input("1. View all \n2. View priority \n ")
    if view_i == "1":
      for row in organizer_list:
        for item in row:
          print(f"{item:^10}, end="|"")
        print()
      print()
    elif view_i == "2":
      prio_i = input(" Enter the priority (high\medium\low) >  ").strip().lower()
      for row in organizer_list:
        if prio_i in row:
          print(row)
    else:
      print(" You provided the wrong input ")
      
  elif user_i == "edit":
    edit_i = input("What do you want to edit? >")
    for row in organizer_list:
      if edit_i in row:
        organizer_list.remove(row)
        task = input("What is the task? >")
        due = input("When is it due by? >")
        priority = input("What is the priority? >")
        row = [task, due, priority]
        organizer_list.append(row)
        print()
        print("Thanks, this task has been added.")
  elif user_i == "remove":
    remove_i = input("What do you want to remove? >")
    for row in organizer_list:
      if remove_i in row:
        organizer_list.remove(row)
        print()
        print("Thanks, this task has been removed.")
  else:
    print("You provided the wron input")
  
    
  x = input("Do you want to quit(y/n) ? ").strip().lower()
  if x[0] == "y":
    break
```
